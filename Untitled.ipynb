```python
from graphviz import Digraph

# Function to create a Context-Level Diagram
def create_context_level_dfd():
    context_diagram = Digraph("Context_Level_DFD", format="png")
    context_diagram.attr(rankdir="LR")

    # External entities
    context_diagram.node("Customer", shape="ellipse")
    context_diagram.node("Admin", shape="ellipse")

    # Process (System)
    context_diagram.node("Website", shape="circle")

    # Data flows
    context_diagram.edge("Customer", "Website", "Browsing requests, Orders")
    context_diagram.edge("Website", "Customer", "Product details, Order confirmations")
    context_diagram.edge("Admin", "Website", "Content updates, Inventory updates")
    context_diagram.edge("Website", "Admin", "Feedback, Customer inquiries")

    # Save and render
    context_diagram.render("context_level_dfd", cleanup=True)

# Function to create a D1-Level Diagram
def create_d1_level_dfd():
    d1_diagram = Digraph("D1_Level_DFD", format="png")
    d1_diagram.attr(rankdir="LR")

    # Processes within the website
    d1_diagram.node("Browse Menu", shape="circle")
    d1_diagram.node("Blog Section", shape="circle")
    d1_diagram.node("About Page", shape="circle")
    d1_diagram.node("Contact Page", shape="circle")

    # External entities
    d1_diagram.node("Customer", shape="ellipse")
    d1_diagram.node("Admin", shape="ellipse")

    # Data stores
    d1_diagram.node("Product Database", shape="cylinder")
    d1_diagram.node("Blog Database", shape="cylinder")
    d1_diagram.node("Contact Database", shape="cylinder")

    # Data flows
    d1_diagram.edge("Customer", "Browse Menu", "View menu")
    d1_diagram.edge("Browse Menu", "Product Database", "Retrieve products")
    d1_diagram.edge("Product Database", "Browse Menu", "Product details")
    d1_diagram.edge("Browse Menu", "Customer", "Menu details")

    d1_diagram.edge("Customer", "Blog Section", "Read blogs")
    d1_diagram.edge("Blog Section", "Blog Database", "Retrieve blog content")
    d1_diagram.edge("Blog Database", "Blog Section", "Blog posts")
    d1_diagram.edge("Blog Section", "Customer", "Blog content")

    d1_diagram.edge("Customer", "About Page", "View about info")
    d1_diagram.edge("About Page", "Customer", "About details")

    d1_diagram.edge("Customer", "Contact Page", "Submit inquiry")
    d1_diagram.edge("Contact Page", "Contact Database", "Store inquiry")
    d1_diagram.edge("Contact Page", "Admin", "Forward inquiries")
    d1_diagram.edge("Admin", "Contact Page", "Respond to inquiries")

    d1_diagram.edge("Admin", "Product Database", "Update products")
    d1_diagram.edge("Admin", "Blog Database", "Update blog content")

    # Save and render
    d1_diagram.render("d1_level_dfd", cleanup=True)

# Generate the diagrams
create_context_level_dfd()
create_d1_level_dfd()
```
